```python


def ask_question(question, tone="Company"):
    """
    Ask a question to the ChatGPT model in a specific tone using OpenAI's API v1.0.0 or newer.

    Parameters:
    - question (str): The question you want to ask.
    - tone (str): The tone of voice for the answer. Options: "friendly", "professional", "neutral", "sarcastic".

    Returns:
    - str: The ChatGPT's response in the specified tone.
    """

    # Define the prompt with the desired tone
    prompt = f"User: {question}\nAI: {tone}\n"

    # Using the updated API call format for chat completions
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "user", "content": question},
            {"role": "assistant", "content": tone}
        ],
        temperature=0  # Adjust as needed
    )

    # Extracting the text from the response object
    #answer = completion.choices[0].message['content'].strip()
    #answer = completion['choices'][0]['message']['content'].strip()
    #answer = completion.messages[1].content.strip()
    answer = completion.choices[0].message.content

    return answer


```
